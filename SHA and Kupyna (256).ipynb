```python
initial_values = [
    '6a09e667', 'bb67ae85', '3c6ef372', 'a54ff53a',
    '510e527f', '9b05688c', '1f83d9ab', '5be0cd19'
]

sha_256_const = [
    '428a2f98', '71374491', 'b5c0fbcf', 'e9b5dba5',
    '3956c25b', '59f111f1', '923f82a4', 'ab1c5ed5',
    'd807aa98', '12835b01', '243185be', '550c7dc3',
    '72be5d74', '80deb1fe', '9bdc06a7', 'c19bf174',
    'e49b69c1', 'efbe4786', '0fc19dc6', '240ca1cc',
    '2de92c6f', '4a7484aa', '5cb0a9dc', '76f988da',
    '983e5152', 'a831c66d', 'b00327c8', 'bf597fc7',
    'c6e00bf3', 'd5a79147', '06ca6351', '14292967',
    '27b70a85', '2e1b2138', '4d2c6dfc', '53380d13',
    '650a7354', '766a0abb', '81c2c92e', '92722c85',
    'a2bfe8a1', 'a81a664b', 'c24b8b70', 'c76c51a3',
    'd192e819', 'd6990624', 'f40e3585', '106aa070',
    '19a4c116', '1e376c08', '2748774c', '34b0bcb5',
    '391c0cb3', '4ed8aa4a', '5b9cca4f', '682e6ff3',
    '748f82ee', '78a5636f', '84c87814', '8cc70208',
    '90befffa', 'a4506ceb', 'bef9a3f7', 'c67178f2'
]


def to_bin(dec):
    return (str(format(dec, 'b')))


def to_bin8(dec):
    return (str(format(dec, '08b')))


def to_bin32(dec):
    return (str(format(dec, '032b')))


def to_bin64(dec):
    return (str(format(dec, '064b')))


def to_hex(dec):
    return (str(format(dec, 'x')))


def bin_to_dec(bin_string):  # 2
    return (int(bin_string, 2))


def hex_to_dec(hex_string):  # 16
    return (int(hex_string, 16))


def string_to_list(bit_string):  # strin to list
    bit_list = []
    for i in range(len(bit_string)):
        bit_list.append(bit_string[i])
    return (bit_list)


def list_to_string(bit_list):  # list to string
    bit_string = ''
    for i in range(len(bit_list)):
        bit_string += bit_list[i]
    return (bit_string)


def rot_r(bit_string, n):  # rotation
    bit_list = string_to_list(bit_string)
    count = 0
    while count <= n - 1:
        list_main = list(bit_list)
        var_0 = list_main.pop(-1)
        list_main = list([var_0] + list_main)
        bit_list = list(list_main)
        count += 1
    return (list_to_string(list_main))


def shift_r(bit_string, n):  # shift
    bit_list = string_to_list(bit_string)
    count = 0
    while count <= n - 1:
        bit_list.pop(-1)
        count += 1
    front_append = ['0'] * n
    return (list_to_string(front_append + bit_list))


def add_by_32_mod(input_set):  # addition by mod 2^32
    value = 0
    for i in range(len(input_set)):
        value += input_set[i]
    mod_32 = 4294967296
    return (value % mod_32)


def xor(bit_string_1, bit_string_2):
    xor_list = []
    for i in range(len(bit_string_1)):
        if bit_string_1[i] == '0' and bit_string_2[i] == '0':
            xor_list.append('0')
        if bit_string_1[i] == '1' and bit_string_2[i] == '1':
            xor_list.append('0')
        if bit_string_1[i] == '0' and bit_string_2[i] == '1':
            xor_list.append('1')
        if bit_string_1[i] == '1' and bit_string_2[i] == '0':
            xor_list.append('1')
    return (list_to_string(xor_list))


def and_op(bit_string_1, bit_string_2):
    and_list = []
    for i in range(len(bit_string_1)):
        if bit_string_1[i] == '1' and bit_string_2[i] == '1':
            and_list.append('1')
        else:
            and_list.append('0')

    return (list_to_string(and_list))


def or_op(bit_string_1, bit_string_2):
    or_list = []
    for i in range(len(bit_string_1)):
        if bit_string_1[i] == '0' and bit_string_2[i] == '0':
            or_list.append('0')
        else:
            or_list.append('1')
    return (list_to_string(or_list))


def not_op(bit_string):
    not_list = []
    for i in range(len(bit_string)):
        if bit_string[i] == '0':
            not_list.append('1')
        else:
            not_list.append('0')
    return (list_to_string(not_list))


# compression
def ch(x, y, z):
    return (xor(and_op(x, y), and_op(not_op(x), z)))


def maj(x, y, z):
    return (xor(xor(and_op(x, y), and_op(x, z)), and_op(y, z)))


def e_0(x):
    return (xor(xor(rot_r(x, 2), rot_r(x, 13)), rot_r(x, 22)))


def e_1(x):
    return (xor(xor(rot_r(x, 6), rot_r(x, 11)), rot_r(x, 25)))


def s_0(x):
    return (xor(xor(rot_r(x, 7), rot_r(x, 18)), shift_r(x, 3)))


def s_1(x):
    return (xor(xor(rot_r(x, 17), rot_r(x, 19)), shift_r(x, 10)))


def pre_proc(message):  # initial processing
    bit_list = []
    for i in range(len(message)):
        bit_list.append(to_bin8(ord(message[i])))  # translating the message into binary
    bit_main = list_to_string(bit_list)
    pd_one = bit_main + '1'
    pd_len = len(pd_one)
    k = 0
    while ((pd_len + k) - 448) % 512 != 0:
        k += 1
    back_ap_0 = '0' * k  # fill everything with zeros without the last 64 bits
    back_ap_1 = to_bin64(len(bit_main))  # and add 64 bits that mean the length of the input message
    pd = pd_one + back_ap_0 + back_ap_1

    to_get = []
    j = 0
    k = 32
    while k < len(pd) + 1:  # divide into 32 bits
        to_get.append(pd[j:k])
        j = k
        k += 32
    return (to_get)


def message_sch(ind, w):
    new_word = to_bin32(add_by_32_mod(
        [int(s_1(w[ind - 2]), 2), int(w[ind - 7], 2), int(s_0(w[ind - 15]), 2), int(w[ind - 16], 2)]))
    return (new_word)


def sha_256(input_string):
    w = pre_proc(input_string)  # create a query
    # initialise variables in 16-system
    a = to_bin32(hex_to_dec(initial_values[0]))
    b = to_bin32(hex_to_dec(initial_values[1]))
    c = to_bin32(hex_to_dec(initial_values[2]))
    d = to_bin32(hex_to_dec(initial_values[3]))
    e = to_bin32(hex_to_dec(initial_values[4]))
    f = to_bin32(hex_to_dec(initial_values[5]))
    g = to_bin32(hex_to_dec(initial_values[6]))
    h = to_bin32(hex_to_dec(initial_values[7]))
    for i in range(0, 64):  # compression cycle
        if i <= 15:
            temp1 = add_by_32_mod([int(h, 2), int(e_1(e), 2), int(ch(e, f, g), 2), int(sha_256_const[i], 16),
                                   int(w[i], 2)])  # h + S1 + ch + k[i] + w[i]
            temp2 = add_by_32_mod([int(e_0(a), 2), int(maj(a, b, c), 2)])  # S0 + maj
            h = g
            g = f
            f = e
            e = add_by_32_mod([int(d, 2), temp1])  # d + temp1
            d = c
            c = b
            b = a
            a = add_by_32_mod([temp1, temp2])  # temp1 + temp2
            a = to_bin32(a)
            e = to_bin32(e)
        if i > 15:
            w.append(message_sch(i, w))
            temp1 = add_by_32_mod(
                [int(h, 2), int(e_1(e), 2), int(ch(e, f, g), 2), int(sha_256_const[i], 16), int(w[i], 2)])
            temp2 = add_by_32_mod([int(e_0(a), 2), int(maj(a, b, c), 2)])
            h = g
            g = f
            f = e
            e = add_by_32_mod([int(d, 2), temp1])
            d = c
            c = b
            b = a
            a = add_by_32_mod([temp1, temp2])
            a = to_bin32(a)
            e = to_bin32(e)
    hash_0 = add_by_32_mod([hex_to_dec(initial_values[0]), int(a, 2)])  # h0 + a
    hash_1 = add_by_32_mod([hex_to_dec(initial_values[1]), int(b, 2)])  # h1 + b ...
    hash_2 = add_by_32_mod([hex_to_dec(initial_values[2]), int(c, 2)])
    hash_3 = add_by_32_mod([hex_to_dec(initial_values[3]), int(d, 2)])
    hash_4 = add_by_32_mod([hex_to_dec(initial_values[4]), int(e, 2)])
    hash_5 = add_by_32_mod([hex_to_dec(initial_values[5]), int(f, 2)])
    hash_6 = add_by_32_mod([hex_to_dec(initial_values[6]), int(g, 2)])
    hash_7 = add_by_32_mod([hex_to_dec(initial_values[7]), int(h, 2)])
    hash = "".join((to_hex(hash_0),
                          to_hex(hash_1),
                          to_hex(hash_2),
                          to_hex(hash_3),
                          to_hex(hash_4),
                          to_hex(hash_5),
                          to_hex(hash_6),
                          to_hex(hash_7)))  # получаем окончательный хеш
    return (hash)
```

```python
def mult_gf(x, y):
    r = 0
    for i in range(8):
        if (y & 0x1) == 1:
            r ^= x
        hbit = x & 0x80
        x = x << 1
        if hbit == 0x80:
            x ^= 0x011d
        y >>= 1
    return r

m_tab = [[0 for j in range(256)] for i in range(256)]
for x in range(256):
        for y in range(256):
            m_tab[x][y] = mult_gf(x, y)

def prepoc_padd(message):
    bytes = [int(message[i * 2: (i + 1) * 2], 16) for i in range(len(message) // 2)]
    binary_message = ''.join(bin(byte)[2:].zfill(8) for byte in bytes)

    N = len(binary_message)
    k = (-97 - N) % l
    binary_N = ''
    for i in range(12):
        binary_N += bin((N >> i * 8) & 0xff)[2:].zfill(8)
    padded_message = binary_message + '1' + '0' * k + binary_N
    stats = []
    for b in range(len(padded_message) // l):
        bits = padded_message[b * l:(b + 1) * l]
        stats.append(int(bits, 2))
    return stats

def xorr(stat, v):
    for j in range(c):
        stat[0][j] ^= (j << 4) ^ v
    return stat

def addd(stat, v):
    for j in range(c):
        word = 0
        for i in range(8):
            word += stat[i][j] << i * 8
        result = word + (0x00f0f0f0f0f0f0f3 ^ ((((c - j - 1) << 4) ^ v) << 56))
        for i in range(8):
            stat[i][j] = (result >> i * 8) & 0xff
    return stat

def sub_byte( stat):
    new_stat = [[0 for j in range(c)] for i in range(8)]
    for i in range(8):
        for j in range(c):
            new_stat[i][j] = sboxes[i % 4][stat[i][j]]
    return new_stat

def shift_byte( stat):
    new_stat = [[0 for j in range(c)] for i in range(8)]
    for i in range(7):
        for j in range(c):
            new_stat[i][(j + i) % c] = stat[i][j]
    for j in range(c):
        shift = 7 if l == 512 else 11
        new_stat[7][(j + shift) % c] = stat[7][j]
    return new_stat

def mix_column( stat):
    new_stat = [[0 for j in range(c)] for i in range(8)]
    for col in range(c):
        for row in range(7, -1, -1):
            product = 0
            for b in range(7, -1, -1):
                product ^= m_tab[stat[b][col]][mds[row][b]]
            new_stat[row][col] = product
    return new_stat

def xor_T( word):
    stat = wrd_to_stat(word)
    for v in range(t):
        stat = xorr(stat, v)
        stat = sub_byte(stat)
        stat = shift_byte(stat)
        stat = mix_column(stat)
    return stat_to_wrd(stat)

def add_T( word):
    stat = wrd_to_stat(word)
    for v in range(t):
        stat = addd(stat, v)
        stat = sub_byte(stat)
        stat = shift_byte(stat)
        stat = mix_column(stat)
    return stat_to_wrd(stat)

def wrd_to_stat( word):
    stat = [[0 for j in range(c)] for i in range(8)]
    hex_word = hex(word)[2:].zfill(l // 4)
    bytes = [int(hex_word[i * 2:(i + 1) * 2], 16) for i in range(len(hex_word) // 2)]
    for j in range(c):
        for i in range(8):
            stat[i][j] = bytes[j * 8 + i]
    return stat

def stat_to_wrd(stat):
    word = 0
    for j in range(c):
        for i in range(8):
            word += stat[i][j] << (c * 8 - 1 - j * 8 - i) * 8
    return word

def kupyna_256_hash( message):
    if l == 512:
        hash_value = 1 << 510
    else:
        hash_value = 1 << 1023
    input_stats = prepoc_padd(message)
    for input_stat in input_stats:
        hash_value = xor_T(hash_value ^ input_stat) ^ add_T(input_stat) ^ hash_value
    return (xor_T(hash_value) ^ hash_value) & int('1' * 256, 2)
```
